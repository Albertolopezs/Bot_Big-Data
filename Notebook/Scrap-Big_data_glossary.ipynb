```python
import time
import sys
import platform
import os
import pandas as pd
import json

from selenium import webdriver
from selenium.common.exceptions import TimeoutException
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.ui import WebDriverWait


driver = None

def load_driver(url):
    """ Logging into our own profile """

    try:
        global driver

        options = Options()

        #  Code to disable notifications pop up of Chrome Browser
        options.add_argument("--disable-notifications")
        options.add_argument("--disable-infobars")
        options.add_argument("--mute-audio")
        # options.add_argument("headless")

        try:
            platform_ = platform.system().lower()
            print(platform_)
            if platform_ in ['linux', 'darwin']:
                print(webdriver.Chrome(executable_path="./chromedriver", options=options))
                driver = webdriver.Chrome(executable_path="./chromedriver", options=options)
            else:
                driver = webdriver.Chrome(executable_path="./chromedriver.exe", options=options)
        except:
            print("Kindly replace the Chrome Web Driver with the latest one from"
                  "http://chromedriver.chromium.org/downloads"
                  "\nYour OS: {}".format(platform_)
                 )
            exit()

        driver.get(url)
        driver.maximize_window()

    except Exception as e:
        print("There's some error in log in.")
        print(sys.exc_info()[0])
        print(e)
        exit()
```

```python
def big_data_glosary():
    global words,descriptions,url_word,url_attachment
    url = "https://bigdata-madesimple.com/big-data-a-to-zz-a-glossary-of-big-data-terminology/"
    load_driver(url)
    
    word = driver.find_elements_by_xpath('(//div[contains(@class,"entry-content")]//h3//strong)')
    print(len(word))
    description = driver.find_elements_by_xpath('//div[contains(@class,"entry-content")]/p')
    for i in range(len(word)-1):
        if(i >=11):
            words.append(word[i+1].text)  
        else:
            words.append(word[i].text)   
        print('Word:',words[-1])
        descriptions.append(description[i+1].text)  
        url_word.append( driver.current_url)
        url_attachment.append([])
    
```

```python
def machine_learning_glosary():
    global words,descriptions,url_word,url_attachment
    url = 'https://www.analyticsvidhya.com/glossary-of-common-statistics-and-machine-learning-terms/'
    load_driver(url)

    tables =driver.find_elements_by_xpath('//div[contains(@class,"su-table")]')
    print('Tablas: ', len(tables))
    description = driver.find_elements_by_xpath('//div[contains(@class,"entry-content")]/p')
    for i in range(1,len(tables)):
        sub_tables = driver.find_elements_by_xpath('//div[contains(@class,"su-table")]['+str(i)+']/table')
        print('Sub-Tablas: ', len(sub_tables))
        for j in range(1,len(sub_tables)+1):
                list_words = driver.find_elements_by_xpath('//div[contains(@class,"su-table")]['+str(i)+']/table['+str(j)+']/tbody/tr')
                print('Lista palabras: ', len(list_words))
                for k in range(2,len(list_words)+1):
                    """Word"""
                    row =  driver.find_elements_by_xpath('//div[contains(@class,"su-table")]['+str(i)+']/table['+str(j)+']/tbody/tr['+str(k)+']/td')
                    word = row[0].text
                    words.append(word)
                    description = row[1].text
                    descriptions.append(description)               
                    print('Word:',word)
                    
                    """Attachments_url"""
                    
                    attachments = driver.find_elements_by_xpath('//div[contains(@class,"su-table")]['+str(i)+']/table['+str(j)+']/tbody/tr['+str(k)+']/td//img')
                    attachments_url = []
                    for attach in attachments:
                        attachments_url.append(attach.get_attribute("src")) 
                    
                    url_attachment.append(attachments_url)
                    url_word.append( driver.current_url)
                
```

```python
def big_data_analitics_glosary():
    global words,descriptions,url_word,url_attachment
    url = " https://www.exasol.com/en/insights/big-data-and-analytics-glossary/"
    load_driver(url)
    word = driver.find_elements_by_xpath('//div[contains(@class,"et_pb")]//p//a')
    print(len(word))
    for i in range(len(word)-1):
        word = driver.find_elements_by_xpath('//div[contains(@class,"et_pb")]//p//a')
        link = word[i].get_attribute("href")
        driver.get(link)
        time.sleep(3)
        question = driver.find_elements_by_xpath('//div[contains(@class,"et_pb_section_1")]//h2')
        answer = driver.find_elements_by_xpath('//div[contains(@class,"et_pb_section_1")]//p')
        images = driver.find_elements_by_xpath('//div[contains(@class,"et_pb_section_1")]//img')
        q_tmp = [ ]
        a_tmp = []
        img_tmp = []
        for j in range(len(question)):
            q_tmp.append(question[j].text)
            a_tmp.append(answer[j].text)
            for k in range(len(images)):
                img_link = images[k].get_attribute("src")
                img_tmp.append(img_link)
        driver.back()
            
        
        words.append(q_tmp)  
        descriptions.append(a_tmp)
        url_attachment.append(img_tmp)
        url_word.append(link)
    
```

```python

def gartner_it_glosary():
    global words,descriptions,url_word,url_attachment
    url = "https://www.gartner.com/it-glossary/"
    load_driver(url)
    _letras = driver.find_elements_by_xpath("//div[contains(@class,'btn')]/a")

    for i in range(len(_letras)): 
        _letras = driver.find_elements_by_xpath("//div[contains(@class,'btn')]/a")
        print(_letras[i].text)
        #Entramos en cada letra
        url_letra = _letras[i].get_attribute("href")
        driver.get(url_letra)
        _word = driver.find_elements_by_xpath("//div[@class='browse-list']//a")
        for j in range(len(_word)):
            _word = driver.find_elements_by_xpath("//div[@class='browse-list']//a")
            _url_palabra = _word[j].get_attribute("href")
            driver.get(_url_palabra)
            _title = driver.find_element_by_xpath("//h2")
            _description = driver.find_element_by_xpath("//div[@class='summary']")
            words.append(_title.text)
            descriptions.append(_description.text)
            url_word.append( driver.current_url)
            url_attachment.append([])
            driver.back()
        driver.back()
            
    
```
