```python
import time
import sys
import platform
import os
import pandas as pd
import json


from selenium import webdriver
from selenium.common.exceptions import TimeoutException
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.ui import WebDriverWait


driver = None

def load_driver(url):
    """ Logging into our own profile """

    try:
        global driver

        options = Options()

        #  Code to disable notifications pop up of Chrome Browser
        options.add_argument("--disable-notifications")
        options.add_argument("--disable-infobars")
        options.add_argument("--mute-audio")
        # options.add_argument("headless")

        try:
            platform_ = platform.system().lower()
            print(platform_)
            if platform_ in ['linux', 'darwin']:
                print(webdriver.Chrome(executable_path="./chromedriver", options=options))
                driver = webdriver.Chrome(executable_path="./chromedriver", options=options)
            else:
                driver = webdriver.Chrome(executable_path="./chromedriver.exe", options=options)
        except:
            print("Kindly replace the Chrome Web Driver with the latest one from"
                  "http://chromedriver.chromium.org/downloads"
                  "\nYour OS: {}".format(platform_)
                 )
            exit()

        driver.get(url)
        driver.maximize_window()

    except Exception as e:
        print("There's some error in log in.")
        print(sys.exc_info()[0])
        print(e)
        exit()
        
with open('../datasets/glossary_faq.json', 'r') as json_file:
    json_file = json.load(json_file)
words_glossary = []
for i in range(len(json_file)):
    try:
        words_glossary.append(json_file[str(i)]['Word'])
    except:
        print("Arreglar json")
```

```python
def wikipedia_glossary():
    url = "https://en.wikipedia.org/wiki/Main_Page"
    load_driver(url)
    text_t = []
    url_word = []
    url_attachment_t = []
    cnt = 0
    for word in words_glossary:
        cnt +=1
        driver.get(url)
        inputElement = driver.find_element_by_xpath("//input[contains(@type,'search')]")
        submitElement = driver.find_element_by_xpath("//input[contains(@type,'submit')]")
        inputElement.send_keys(word)
        retryingFindClick(submitElement)
        try:
            body_text = driver.find_elements_by_xpath('//div[contains(@id,"content-text")]//p')
            text = ""
            for p in body_text:
                if(p.text != ""):
                    text += p.text
            text_t.append(text)
            url_word.append( driver.current_url)
                ###Attachments
            attachments = driver.find_elements_by_xpath("//div[contains(@id, 'bodyContent') and contains(@class, 'body-content')]//img")
            attachments_url = []
            for attach in attachments:
                attachments_url.append(attach.get_attribute("src")) 
                url_attachment_t.append(attachments_url)
        except:
            text_t.append([])
            url_word.append([]) 
            url_attachment_t.append([])
            print("No se encuentra la p√°gina",cnt)
    
    return text_t,url_word,url_attachment_t

#Esto es porque a veces el DOM no se carga a tiempo
def retryingFindClick(elemento):
    result = False
    attempts = 0
    while(attempts < 2) :
        try:
            elemento.click()
            result = True
            break;
        except:
        
            attempts += 1

    return result
```

```python
text_t,url_word,url_attachment_t = wikipedia_glossary()
```

```python
json_total = {}
for x in range(len(words_glossary)):
    json_d = {'Word':words_glossary[x],'Text':text_t[x],'Url':url_word[x],'Attachment_Url':url_attachment_t[x]}
    json_total[x] = json_d
    json_total.update(json_d)


with open('../datasets/wikipedia_glossary.json', 'w') as outfile:
    json.dump(json_total, outfile)
```
